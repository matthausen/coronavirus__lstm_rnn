```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

import geopandas
```

```python
df = pd.read_csv('data/covid_march.csv')
covid = df.groupby(df['Country/Region']).sum().reset_index()
covid.head()
```

```python
states_shapes = geopandas.read_file('data/Countries_WGS84/Countries_WGS84.shp')
states_shapes.rename(columns={'CNTRY_NAME':'Country/Region'}, inplace=True)
states_shapes.head()
```

```python
states_shapes=states_shapes.merge(covid, on='Country/Region', how='left')
states_shapes.head()
```
